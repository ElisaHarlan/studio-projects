```python
df_Stuttgart['ratio'] = df_Stuttgart['Personal der Pflegeheime'] / df_Stuttgart['Verfügbare Plätze in Pflegeheimen']
```

```python
df_Stuttgart
```

```python
#df_Stuttgart[['Verfügbare Plätze in Pflegeheimen', 'Personal der Pflegeheime']].pct_change().plot(kind='barh',ylim=(0,0.10))
```

```python
#df_Stuttgart[['Pflegedienste', 'Personal der Pflegedienste']].pct_change().plot(kind='barh',ylim=(0,0.10))
```

```python
ax = df_Stuttgart['Pflegeheime'].plot(color='red', legend=True, ylim=(0,80))
df_Stuttgart['Pflegedienste'].plot(color='blue', ax=ax, legend=True)
ax.invert_xaxis()
```

```python
ax = df_Stuttgart['Pflegeheime'].pct_change().dropna().plot(ylim=(-0.25,0.10), color='red', legend=True)
df_Stuttgart['Pflegedienste'].pct_change().dropna().plot(ylim=(-0.25,0.10), color='blue', legend=True)
ax.invert_xaxis()
```

```python
df_Stuttgart['Pflegeheime_Prozent'] = \
(df_Stuttgart['Pflegeheime']/df_Stuttgart['Pflegeheime'].iloc[-1]-1)*100
```
