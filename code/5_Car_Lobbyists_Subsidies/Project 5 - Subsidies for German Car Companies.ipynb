```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
```

```python
df = pd.read_excel('Final.xlsx')
```

```python
df
```

```python
df['2008'].fillna(0, inplace=True)
df['2009'].fillna(0, inplace=True)
df['2010'].fillna(0, inplace=True)
df['2011'].fillna(0, inplace=True)
df['2012'].fillna(0, inplace=True)
```

```python
df['Total'] = df['2007'] + df['2008'] + df['2009'] + df['2010'] + df['2011'] + df['2012'] + df['2013'] + df['2014'] + df['2015'] + df['2016']
```

```python
# Import 
from matplotlib.ticker import FormatStrFormatter

# Plot and save it to an ax
ax = df.sort_values("Total", ascending=False)["Total"].head().plot(y='Total', x='Company', kind='bar', label='Company')

# Change the y axis formatting
ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
```

```python
import matplotlib.pyplot as plt    
plt.style.use('ggplot')
colors = ['b', 'green', 'y', 'pink','orange','cyan','darkgrey']
ax = df.sort_values("Total", ascending=False)["Total"].head(10).plot(y='Total', x='Company', kind='bar', color=colors, legend=False, label='Company')
ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))

for bar in ax.patches:
    bar.set_facecolor('#888888')

highlight = 'Robert Bosch GmbH'
pos = df.index.get_loc(highlight)

ax.patches[pos].set_facecolor('#aa3333')

ax.set_xlabel("Company",fontsize=12)
ax.set_ylabel("Total",fontsize=12)
ax.set_xticklabels(["Robert Bosch GmbH", "Siemens AG","Siemens Aktiengesellschaft", "Daimler AG", "Robert Bosch Gesellschaft", "Daimler AG", "Daimler AG", "VW AG", "Volkswagen Aktiengesellschaft", "Siemens AG"])  
ax.legend()
```
